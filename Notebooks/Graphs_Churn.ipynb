```python
import chardet
import pandas as pd

def import_dataset(Loc):
    # Read the raw data to detect encoding
    rawdata = open(Loc, 'rb').read()
    result = chardet.detect(rawdata)

    # Read the CSV file using the detected encoding
    data = pd.read_csv(Loc, encoding=result['encoding'])

    return data
```

```python
churn_results=import_dataset(r'/content/churn_results.csv')
churn_results['Balancing Technique'].fillna('Standard', inplace=True)
```
